```python
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
names=['sepal length','sepal-width','petal length','petal width','class']
df=pd.read_csv(url,names=names)
df.head(10)
```

```python
x=df.iloc[:,0].values
y=df.iloc[:,1].values
validation_points=0.20
seed=6
#print(y)
x1=df.iloc[:,2].values
y1=df.iloc[:,3].values
x4=df.iloc[:,4].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=validation_points,random_state=seed)
```

# #Plotting graphs

simple line graphs

```python
import matplotlib.style as stl
stl.use('seaborn-deep')
#plt.plot(x,x1,'g',label='sepal and petal length',linewidth=2)
#plt.plot(y,y1,'r',label='sepal and petal width',linewidth=2)
plt.plot(x4,x,linewidth=2,color='g',label='sepal length')
plt.plot(x4,y,linewidth=2,color='c',label='sepal width')
plt.plot(x4,x1,linewidth=2,color='y',label='petal length')
plt.plot(x4,y1,linewidth=2,color='r',label='petal width')
plt.title('length and width of the sepal and petal')
plt.xlabel('class')
plt.ylabel('length  & width')
plt.legend()
#plt.grid(True,color='k')
plt.show()
```

Bar graphs

```python

stl.use('dark_background')
width=0.3
plt.bar(x4,x,width,color='g',label='sepal length')
plt.bar(x4,y,width,color='c',label='sepal width')
plt.bar(x4,x1,width,color='y',label='petal length')
plt.bar(x4,y1,width,color='r',label='petal width')
plt.title('length and width of the sepal and petal')
plt.xlabel('class')
plt.ylabel('length and width')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

Histogram

```python
bins=np.arange(50)
plt.hist(x,bins,color='g',histtype='bar',width=1)
plt.title('length and width of the sepal and petal')
plt.xlabel('class')
plt.ylabel('length and width')
#plt.legend()
plt.grid(True,color='y')
plt.show()
```

Scatter plot

```python
stl.use('seaborn-dark')
width=29
plt.scatter(x,y,width,color='b',label='sepal length & width')
plt.scatter(x1,y1,width,color='r',label='petallength & width')
#plt.scatter(x4,x1,width,color='y',label='petal length')
#plt.scatter(x4,y1,width,color='r',label='petal width')
plt.title('length and width of the sepal and petal')
plt.xlabel('sepal and petal length')
plt.ylabel('sepal and petal width')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

Pie Chart

```python
slices=[x.sum(),y.sum(),x1.sum(),y1.sum()]
names=['sepal length','sepal-width','petal length','petal width']
color=['c','g','r','b']
plt.pie(slices,
       labels=names,
       colors=color,
       shadow=True,
       explode=(0,0,0.1,0),
       autopct='%1.1f%%',startangle=90)
plt.title('pie chart')
plt.show()
```
