```python
import matplotlib.pyplot as plt
import matplotlib.style as stl
#https://python-graph-gallery.com/199-matplotlib-style-sheets/
```

```python
x=[1,2,3,6]
y=[9,4,3,8]
x1=[9,4,5]
y1=[4,3,7]
```

Simple graph

```python
#stl.use('dark_background')
style_types = ['seaborn-darkgrid','seaborn-notebook', 
                'classic','seaborn-ticks','grayscale',
                'bmh',  'seaborn-talk','dark_background', 
                'ggplot', 'fivethirtyeight','_classic_test', 
                'seaborn-colorblind','seaborn-deep', 
                'seaborn-whitegrid','seaborn-bright', 
                'seaborn-poster','seaborn-muted', 
                'seaborn-paper','seaborn-white', 
                'seaborn-pastel','seaborn-dark','seaborn', 
                'seaborn-dark-palette']

#for i,j in enumerate(style_types):
 #   print(i,j)
```

```python
stl.use('seaborn-darkgrid')
plt.plot(x,y,'g',label='first line',linewidth=5)
plt.plot(x1,y1,'c',label='second line',linewidth=4)
plt.title('graph')
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

Bar Graph

```python
stl.use('seaborn-dark')
plt.bar(x,y,color='g',label='first line',width=0.6)
plt.bar(x1,y1,color='y',label='second line',width=.6)
plt.legend()
plt.title('Bar graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True,color='r')
plt.show()
```

histogram

```python
bins=[1,2,3,4,5]
plt.hist(x,bins,color='g',histtype='bar',rwidth=0.5)
plt.hist(y,bins,color='c',histtype='bar',rwidth=0.6)
#plt.legend()
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True,color='r')
plt.show()
```

Scatter plot

```python
width=200
plt.scatter(x,y,width,label='first points',color='r')
plt.scatter(x1,y1,width,label='second points',color='b')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Pie Chart

```python
slices=[79,67,37,7]
names=['first points','2nd points','3rd points','4th points']
color=['r','b','g','y']
plt.pie(slices,labels=names,colors=color,shadow=True,explode=(0,0.2,0,0),autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
